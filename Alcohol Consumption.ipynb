## Alcohol Consumption

Source: https://www.gapminder.org/

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
#Read the data
data = pd.read_csv('gapminder_alcohol.csv',sep=',')
```

```python
data
```

```python
#Turn it into a DataFrame
df = pd.DataFrame(data)
```

```python
df
```

```python
#Rename the columns
df = df.rename(columns={'country':'Country','alcconsumption':'Alcohol Consumption','incomeperperson':'Income per Person',
                   'suicideper100th':'Suicide per 100th','employrate':'Employment Rate','urbanrate':'Urban Rate'})
```

```python
df
```

```python
#Remove lines with NaN
df.dropna(axis=0,inplace = True)
```

```python
df
```

```python
#Rebuild new index
df.index = range(df.shape[0])
```

```python
df
```

```python
#Round the numbers in Suicide per 100th
df.loc[:,'Suicide per 100th'] = df.loc[:,'Suicide per 100th'].round(2)
```

```python
df
```

```python
#Do the same for Employment Rate
df.loc[:,'Employment Rate'] = df.loc[:,'Employment Rate'].round(2)
```

```python
df
```

```python
#And for Income per Person
df.loc[:,'Income per Person'] = df.loc[:,'Income per Person'].round(2)
```

```python
df
```

#### Charts

```python
#Alcohol Consumption vs Income
x = df['Alcohol Consumption']
y = df['Income per Person']

plt.rc('figure',figsize=(15,10))
plt.xlabel('Alcohol Consumption (liters)',{'fontsize':10})
plt.ylabel('Income per Person $',{'fontsize':10})
plt.title('Alcohol Consumption x Income per Person',{'fontsize':15})

plt.scatter(x,y,s = 50)
```

```python
#Alcohol Consumption vs Suicide per 100th
x = df['Alcohol Consumption']
y1 = df['Suicide per 100th']

plt.rc('figure',figsize=(15,10))
plt.xlabel('Alcohol Consumption (liters)',{'fontsize':10})
plt.ylabel('Suicide per 100th',{'fontsize':10})
plt.title('Alcohol Consumption x Suicide per 100th',{'fontsize':15})

plt.scatter(x,y1,s=50,c='#7f7f7f')
```

### Identifying and removing Outliers

```python
#Identifying outliers in Alcohol Consumption
df_cons = pd.DataFrame(df['Alcohol Consumption'])
```

```python
df_cons
```

```python
#Generates boxplot
#df_cons.boxplot(['Alcohol Consumption'])
x1 = df_cons['Alcohol Consumption']
plt.title('Boxplot',{'fontsize':15})
plt.boxplot(x1)
```

```python
alcohol = df['Alcohol Consumption']
```

```python
#First quartile
Q1 = alcohol.quantile(0.25)

#Third quartile
Q3 = alcohol.quantile(0.75)

#Inter-quartile range (IQR)
IQR = Q3-Q1
```

```python
IQR
```

```python
#lower whisker boundary
lower = Q1 - 1.5*IQR

#Upper whisker boundary
upper = Q3 + 1.5*IQR
```

```python
#Boxplot without outliers
selection = (alcohol >= lower) & (alcohol<=upper)
new_df_cons = df_cons[selection]
```

```python
new_df_cons
```

```python
x2 = new_df_cons['Alcohol Consumption']
plt.title('Boxplot',{'fontsize':15})
plt.boxplot(x2)
```
