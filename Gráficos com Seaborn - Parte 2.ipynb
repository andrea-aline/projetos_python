```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
choco_data = pd.read_csv('chocolate.csv',index_col='ref',parse_dates=True)
```

```python
choco_data.head()
```

```python
#Confere qual é o tipo de dado das colunas count_of_ingredients
choco_data['counts_of_ingredients']
```

```python
#O mesmo para a coluna rating
choco_data['rating']
```

## Gráficos de Dispersão

```python
#Faz um gráfico de dispersão
sns.scatterplot(x=choco_data['counts_of_ingredients'],y=choco_data['rating'])
plt.title('Count of Ingredients x Rating')
plt.xlabel('Count of Ingredients')
plt.ylabel('Rating')
plt.figure(figsize=(10,15))
```

```python
#Confere o tipo de dados da coluna cocoa_percent
choco_data['cocoa_percent']
```

```python
#Faz outro gráfico de dispersão
sns.set_palette('bright')
sns.scatterplot(x=choco_data['cocoa_percent'],y=choco_data['rating'])
plt.title('Cocoa Percent x Rating')
plt.xlabel('Cocoa Percent')
plt.ylabel('Rating')
plt.figure(figsize=(10,15))
```

```python
#Adiciona uma linha de regressão
sns.regplot(x=choco_data['cocoa_percent'],y=choco_data['rating'])
plt.title('Cocoa Percent x Rating')
plt.xlabel('Cocoa Percent')
plt.ylabel('Rating')
plt.figure(figsize=(10,15))
```

```python
#Códigos de cores no gráfico de regressão
#sns.set_palette('bright')
sns.scatterplot(x=choco_data['cocoa_percent'],y=choco_data['rating'],hue=choco_data['sweetener_without_sugar'],palette=['green','red'])
plt.title('Cocoa Percent x Rating & Sweetener')
plt.xlabel('Cocoa Percent')
plt.ylabel('Rating')
plt.legend(loc='lower left')
plt.figure(figsize=(20,20))
```

```python
#Inclui uma variável categórica
sns.swarmplot(x=choco_data['sweetener_without_sugar'],y=choco_data['rating'])
plt.title('Sweetener x Rating')
plt.xlabel('Sweetener')
plt.ylabel('Rating')
plt.figure(figsize=(20,20))
```

```python
#Obedeço ao aviso anterior e diminuo o tamanho dos markers
plt.figure(figsize=(10,5))
sns.swarmplot(x=choco_data['sweetener_without_sugar'],y=choco_data['rating'],size=3)
plt.title('Sweetener x Rating')
plt.xlabel('Sweetener')
plt.ylabel('Rating')
```

```python
#Como o aviso continua, tento com o stripplot
plt.figure(figsize=(10,5))
sns.stripplot(x=choco_data['sweetener_without_sugar'],y=choco_data['rating'])
plt.title('Sweetener x Rating')
plt.xlabel('Sweetener')
plt.ylabel('Rating')
```

## Heatmap

```python
import seaborn.matrix as matrix
```

```python
#Declaro a variável choco_data novamente para remover ref como índice
choco_data = pd.read_csv('chocolate.csv',parse_dates=True)
```

```python
#Cria o DataFrama choco_data agrupado pelas colunas rating e count_of_ingredients
group =  choco_data.groupby(['rating','counts_of_ingredients'])
```

```python
#Faz uma contagem única 
unique_count = group.ref.nunique()
```

```python
unique_count.head()
```

```python
#Faz um "pivot" para que os dados se tornem largos("wide") ao invés de type ("ajustados")
jb_unique_count = unique_count.reset_index().pivot(index='rating',columns='counts_of_ingredients',values='ref')
```

```python
#Exibe até a quinta linha e até a sexta coluna
jb_unique_count.iloc[:5,:6]
```

```python
#Substitui os valores NaN por 0 (zero)
jb_unique_count.fillna(0,inplace=True)
```

```python
#Exibe o valor mínimo da coluna rating
choco_data['rating'].min()
```

```python
#Exibe o valor máximo da coluna rating
choco_data['rating'].max()
```

```python
#Finalmente, faz o heatmap
plt.figure(figsize=(10,5))
plt.title('Count of Ingredients vs Rating')
plt.xlabel('Count of Ingredients')
plt.ylabel('Rating')
sns.heatmap(jb_unique_count)
plt.show()
```

```python
#Mudando a cor
plt.figure(figsize=(10,5))
plt.title('Count of Ingredients vs Rating')
plt.xlabel('Count of Ingredients')
plt.ylabel('Rating')
sns.heatmap(jb_unique_count,cmap='mako')
plt.show()
```

```python
#Ajustando a escala de cor 
plt.figure(figsize=(10,5))
plt.title('Count of Ingredients vs Rating')
plt.xlabel('Count of Ingredients')
plt.ylabel('Rating')
sns.heatmap(jb_unique_count,cmap='mako',vmin=0,vmax=40)
plt.show()
```

```python
#Insere labels em uma parte dos dados e coloca bordas brancas
plt.figure(figsize=(10,5))
plt.title('Count of Ingredients vs Rating')
plt.xlabel('Count of Ingredients')
plt.ylabel('Rating')
sns.heatmap(jb_unique_count.iloc[1:5,1:5],cmap='mako',vmin=0,vmax=40,annot=True,linewidth=1,linecolor='w',square=True)
plt.show()
```
