Ao olhar o arquivo em excel, vi que nas linhas em que os likes estão como strings, a coluna "video_error_or_removed" está vazia

```python
#Mostra as linhas em que video_error_or_removed" está vazia
gb_videos.loc[gb_videos.video_error_or_removed.isnull() == True]
```

```python
#Tento então usando a própria função drop, mas retorna erro
gb_videos = gb_videos.drop(gb_videos[gb_videos.video_error_or_removed.isnull == True].index)
```

```python
#E os valores NaN continuam
gb_videos.loc[gb_videos.video_error_or_removed.isnull() == True]
```

```python
#Assim, tento outra vez converter 'likes' para inteiro
gb_videos['likes'] = pd.to_numeric(gb_videos['likes'])
```

```python
#Tento então usando dropna(subset) para eliminar as linhas nulas na coluna video_error_or_removed	
gb_videos = gb_videos.dropna(subset=['video_error_or_removed'],inplace=False)
```

```python
#Confiro se os valore nulos na coluna video_error_or_removed foram mesmo eliminados
gb_videos.loc[gb_videos.video_error_or_removed.isnull() == True]
```
